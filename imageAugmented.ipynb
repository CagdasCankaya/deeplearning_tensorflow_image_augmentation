```python
import tensorflow as tf #models
import numpy as np #math computing
import matplotlib.pyplot as plt #plotting chart

import tensorflow_datasets as tfds
from tensorflow.keras.layers import Conv2D,MaxPool2D,Dense,Flatten,InputLayer,BatchNormalization,Input,Dropout
from  tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError, FalsePositives, FalseNegatives, TruePositives, TrueNegatives, Precision, Recall, BinaryAccuracy, AUC
from tensorflow.keras.losses import MeanSquaredError,Huber,MeanAbsoluteError , BinaryCrossentropy
from tensorflow.keras.callbacks import Callback , CSVLogger , EarlyStopping, LearningRateScheduler
from tensorflow.keras.regularizers import L2,L1

import sklearn
from sklearn.metrics import confusion_matrix, roc_curve
import seaborn as sns
from google.colab import drive
```

```python
(train_images, train_labels), _ = tf.keras.datasets.mnist.load_data()

# create dataset
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels)).batch(1)

def visualize(original, augmented):
    plt.figure(figsize=(8, 4))

    # show original img
    plt.subplot(1, 2, 1)
    plt.imshow(original[0], cmap='gray')  # take img original[0] numpy array
    plt.title('Original')

    # Augmente edilmiş görüntüyü gösterme
    plt.subplot(1, 2, 2)
    plt.imshow(augmented[0], cmap='gray')  # take img augmented[0] numpy array
    plt.title('Augmented')

    plt.show()

# take next example and show img
original_image, label = next(iter(train_dataset))

# img (flip left-right)
augmented_image = tf.image.flip_left_right(original_image)

# visualize img via function
visualize(original_image, augmented_image)
```
